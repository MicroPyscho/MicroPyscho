#OLS Panel Regression Using Random Effects

```python
from linearmodels.panel import PanelOLS, RandomEffects

# Convert 'startyear' to a date-like format
star_df['startyear'] = pd.to_datetime(star_df['year'], format='%Y')  # Adjust format if needed

# Set index for panel data
star_df.set_index(['year', 'ICBIC'], inplace=True)

# Fixed Effects (FE) Model
fe_model = PanelOLS(dependent=star_df['ai_score'], exog=star_df.drop(columns=['ai_score', 'ai_govt', 'ai_tech', 'ai_infr']), entity_effects=True)
fe_results = fe_model.fit()

# Random Effects (RE) Model
re_model = RandomEffects(dependent=star_df['ai_score'], exog=star_df.drop(columns=['ai_score', 'ai_govt', 'ai_tech', 'ai_infr']))
re_results = re_model.fit(check_rank=False)  # Setting check_rank=False to proceed despite rank deficiency

# Get regression summaries
fe_summary = fe_results.summary.as_text()
re_summary = re_results.summary.as_text()

# Print regression summaries in a tabular format
print("Fixed Effects (FE) Model:")
print(fe_summary)

print("\nRandom Effects (RE) Model:")
print(re_summary)
```

```python
from linearmodels.panel import PanelOLS, RandomEffects
from tabulate import tabulate

# Assuming star_df is your DataFrame containing the variables
# Replace non-numeric values with NaN
star_df_numeric = star_df.apply(pd.to_numeric, errors='coerce')

# Replace infinite values with NaN
star_df_cleaned = star_df_numeric.replace([np.inf, -np.inf], np.nan)

# Replace NaNs with average of non-NaNs in the columns
star_df_cleaned_filled = star_df_cleaned.apply(lambda x: x.fillna(np.average(x.dropna())), axis=0)

# Define independent variables (X) and dependent variable (y) after filling missing values
X = star_df_cleaned_filled[top_correlations.index[1:]]  # Exclude ai_score itself
y = star_df_cleaned_filled['ai_score']

# Fixed Effects (FE) Model
fe_model = PanelOLS(y, X, entity_effects=True)
fe_results = fe_model.fit()

# Random Effects (RE) Model
re_model = RandomEffects(y, X)
re_results = re_model.fit()

# Get regression summaries
fe_summary = fe_results.summary.as_text()
re_summary = re_results.summary.as_text()

# Print regression summaries in a tabular format
print("Fixed Effects (FE) Model:")
print(tabulate(fe_summary, headers='keys', tablefmt='psql'))

print("\nRandom Effects (RE) Model:")
print(tabulate(re_summary, headers='keys', tablefmt='psql'))
```
